```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import gaussian_filter
```

```python
img = cv2.imread('images/AgesOfEmpire.jpeg')
```

```python
img.shape
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_img)
#plt.axis("off")
plt.show()
```

```python
height, width = img.shape[:2]

patch_size = 16

rows = height // patch_size
cols = width // patch_size


patches = []
for row in range(rows):
    for col in range(cols):
        x1 = col * patch_size
        y1 = row * patch_size
        x2 = x1 + patch_size
        y2 = y1 + patch_size

        patch = img[y1:y2, x1:x2]

        patches.append(patch)
```

```python
for i in range(3):
    patch = patches[i]
    patch_norm = (patch - np.mean(patch)) / np.std(patch)
    plt.imshow(patch_norm, cmap='gray')
    plt.axis("off")
    plt.show()
```

```python
def mostrar_imagen(img1,img2,titulo1="original",titulo2="Nueva"):
  plt.figure(figsize=(10,10))
  plt.subplot(1,2,1)
  plt.imshow(cv2.cvtColor(img1,cv2.COLOR_BGR2RGB),cmap='gray')
  #plt.axis("off")
  plt.title(titulo1)
  plt.subplot(1,2,2)
  plt.imshow(cv2.cvtColor(img2,cv2.COLOR_BGR2RGB))
  plt.title(titulo2)
  plt.show()
```

```python
o = 1 / (2 * 0.693)
# Forma de sacar un kernel
# size = 3
# kernel = np.zeros((size, size))
# kernel[size//2, size//2] = 1
# kernel = gaussian_filter(kernel,o)
# kernel = kernel / np.sum(kernel)
```

```python
img2 = cv2.imread('images/AgesOfEmpire.jpeg')
img_res = gaussian_filter(img2, sigma=o,truncate=3.0)
mostrar_imagen(img2, img_res,titulo1='Imagen original',titulo2='Imagen Filtrada')
```

```python
def mostrar_imagen4(img1, img2, img3, img4, titulo1 = 'Mod1', titulo2 = 'Mod2',titulo3 = 'Mod3', titulo4 = 'Mod4'):
  plt.figure(figsize=(10,10))
  plt.subplot (2,2,1)
  plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
  plt.title(titulo1)

  plt.subplot(2,2,2)
  plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
  plt.title(titulo2)


  plt.subplot(2,2,3)
  plt.imshow(cv2.cvtColor(img3, cv2.COLOR_BGR2RGB))
  plt.title(titulo3)

  plt.subplot(2,2,4)
  plt.imshow(cv2.cvtColor(img4, cv2.COLOR_BGR2RGB))
  plt.title(titulo4)

  plt.show()
```
