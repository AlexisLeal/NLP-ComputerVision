```python
def explore_tagset_fd(genero, tagset):
  palabras_etiq = brown.tagged_words(categories=genero, tagset= tagset)
  etq = [tag for word, tag in palabras_etiq]
  frecuencia  = nltk.FreqDist(etq)
  #Obtemos las keys
  num_etiquetas = frecuencia.keys()
  top_etq = frecuencia.most_common(10)
  return num_etiquetas, top_etq
```


### Infomracion muy importante impartida por chapGPT
 El HiddenMarkovModelTagger (Etiquetador de Modelo de Markov Oculto, en español) es un algoritmo de etiquetado gramatical que se utiliza para asignar etiquetas gramaticales a cada palabra en una oración. Este algoritmo se basa en el modelo de Markov oculto (HMM, por sus siglas en inglés), que es una herramienta estadística utilizada para modelar la secuencia de eventos.

En el contexto del etiquetado gramatical, un HMM se utiliza para modelar la probabilidad de que una secuencia de etiquetas gramaticales se asigne a una secuencia de palabras en una oración. El modelo se entrena utilizando un corpus etiquetado, que es un conjunto de oraciones en el que cada palabra ya está etiquetada con su parte del habla correspondiente (como sustantivo, verbo, adjetivo, etc.). A partir de este corpus, el algoritmo aprende la probabilidad de que una etiqueta gramatical siga a otra, así como la probabilidad de que una palabra dada tenga una etiqueta gramatical específica.

Una vez que se entrena el modelo, se puede utilizar para etiquetar nuevas oraciones. Para etiquetar una oración, el algoritmo de HMM selecciona la secuencia de etiquetas gramaticales más probable para la secuencia de palabras de la oración. Este proceso se llama "decodificación" y se realiza utilizando el algoritmo de Viterbi.

En resumen, el HiddenMarkovModelTagger es un algoritmo que utiliza el modelo de Markov oculto para etiquetar oraciones con sus partes del habla correspondientes. Se entrena utilizando un corpus etiquetado y utiliza el algoritmo de Viterbi para seleccionar la secuencia de etiquetas más probable para una oración dada.

```python
def entrenamientoHMM(oracion, tamanio):
    oraciones_etiq = brown.tagged_sents(categories='news')
    #[-tamanio:] selecionar las ultimas palabras(tamaño)
    datos_train = oraciones_etiq[-tamanio:]
    #[:100] seleciona las primeras 100
    datos_test = oraciones_etiq[:100]
    #Entrenamos el modelo
    tagger = HiddenMarkovModelTagger.train(datos_train)
    #etiqueta la oracion a base de su entrenamiento
    oracion_etiquetada_hmm = tagger.tag(oracion)
    # obtiene la precision del modelo
    acc = tagger.evaluate(datos_test)
    return tagger, oracion_etiquetada_hmm, acc
```

```python
def evaluarHMM():
  oraciones_etiq = brown.tagged_sents(categories='news')
  oracion = [tp[0] for tp in oraciones_etiq[76]]
  tagger, oracion_etiquetada_hmm, acc = entrenamientoHMM(oracion, 500)
  print("Entrenando el modulo nltk.HiddenMarkovModelTagger con 500 oraciones")
  print("\tOracion etiquetada con el modelo:")
  pp.pprint(oracion_etiquetada_hmm)
  print("Accuracy del conjunto de pruebas: %.4f%%" % (100.0*acc))
  print()

  tagger, oracion_etiquetada_hmm, acc = entrenamientoHMM(oracion, 3000)
  print("Entrenando el modulo nltk.HiddenMarkovModelTagger con 3000 oraciones")
  print("\tOracion etiquetada con el modelo:")
  pp.pprint(oracion_etiquetada_hmm)
  print("Accuracy del conjunto de pruebas: %.4f%%" % (100.0*acc))
  print()
```

```python
evaluarHMM()
```
