```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('img-destacada-donde-se-aplica-diseno-editorial.jpg',cv.IMREAD_GRAYSCALE)
plt.imshow(img)
```

```python
ver = (cv.__version__).split('.')

if int(ver[0]) < 3.5:
  b_detector = cv.SimpleBlobDetector()
else:
  b_detector = cv.SimpleBlobDetector_create()

keypoints = b_detector.detect(img)
img_con_blobs = cv.drawKeypoints(img, keypoints, np.array([]),(255,255,0), cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.imshow(img_con_blobs)
```
