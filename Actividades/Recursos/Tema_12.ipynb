```python
def mostrar_imagen(img1, img2, titulo1 = 'Original', titulo2 = 'Nueva'):
  plt.figure(figsize=(10,10))
  plt.subplot (1,2,1)
  plt.imshow(cv.cvtColor(img1, cv.COLOR_BGR2RGB))
  plt.title(titulo1)
  plt.subplot(1,2,2)
  plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB))
  plt.title(titulo2)
  plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
```

```python
img = cv. imread('220px-Dante_Alighieri01.jpg')
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
```

```python
renglones, cols,_= img.shape
ruido = np.random.normal(0,15,(renglones, cols, 3)).astype(np.uint8)
img_ruido= img+ruido
mostrar_imagen(img, img_ruido,titulo1='Original',titulo2='Imagen con ruido')
```

```python
from scipy.ndimage import gaussian_filter
img_res = gaussian_filter(img, sigma = 1)
mostrar_imagen(img, img_res,titulo1='Original',titulo2='Imagen con filtrada')
```

```python
img_res =gaussian_filter(img, sigma =  3)
mostrar_imagen(img, img_res,titulo1='Original',titulo2='Imagen con ruido')
```

```python
img_res =gaussian_filter(img, sigma =  7)
mostrar_imagen(img, img_res,titulo1='Original',titulo2='Imagen con ruido')
```

```python
img_res =gaussian_filter(img, sigma =  11)
mostrar_imagen(img, img_res,titulo1='Original',titulo2='Imagen con ruido')
```

```python
img_res =  cv.bilateralFilter(img, 11,120,120)
mostrar_imagen(img, img_res,titulo1='Original',titulo2='Imagen Filtrada')
```

```python
img_1 =gaussian_filter(img, sigma =  3)
img_2 =gaussian_filter(img, sigma =  7)
img_3 =gaussian_filter(img, sigma =  11)
img_4 =gaussian_filter(img, sigma =  15)

def mostrar_imagen4(img1, img2, img3, img4, titulo1 = 'Mod1', titulo2 = 'Mod2',titulo3 = 'Mod3', titulo4 = 'Mod4'):
  plt.figure(figsize=(10,10))
  plt.subplot (2,2,1)
  plt.imshow(cv.cvtColor(img1, cv.COLOR_BGR2RGB))
  plt.title(titulo1)

  plt.subplot(2,2,2)
  plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB))
  plt.title(titulo2)


  plt.subplot(2,2,3)
  plt.imshow(cv.cvtColor(img3, cv.COLOR_BGR2RGB))
  plt.title(titulo3)

  plt.subplot(2,2,4)
  plt.imshow(cv.cvtColor(img4, cv.COLOR_BGR2RGB))
  plt.title(titulo4)

  plt.show()

mostrar_imagen4(img_1, img_2,img_3,img_4,'Img escala 3','Img escala 7','Img escala 11','Img escala 15')
```
