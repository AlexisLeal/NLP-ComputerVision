```python
imag_bf= cv.imread('../Actividad-5/Imagenes/Bf2042.jpg')
```

```python
imag_bf.shape
```

```python
print("Valor maximo",np.max(imag_bf))
print("Valor minimo",np.min(imag_bf))
```

```python
imag_bf = cv.cvtColor(imag_bf, cv.COLOR_BGR2RGB)
plt.axis("off")
plt.imshow(imag_bf)
```

```python
imag_bf = cv.cvtColor(imag_bf, cv.COLOR_BGR2RGB)
plt.imshow(imag_bf, cmap='gray')
```

```python
# Redimensiona las imgenes modificando solo el factor de escala en el eje horizontal a 2
img1_resized_nearest = cv.resize(imag_halo, (imag_halo.shape[1]*2, imag_halo.shape[0]), interpolation=cv.INTER_NEAREST)
img1_resized_cubic = cv.resize(imag_halo, (imag_halo.shape[1]*2, imag_halo.shape[0]), interpolation=cv.INTER_CUBIC)
img2_resized_nearest = cv.resize(imag_bf, (imag_bf.shape[1]*2, imag_bf.shape[0]), interpolation=cv.INTER_NEAREST)
img2_resized_cubic = cv.resize(imag_bf, (imag_bf.shape[1]*2, imag_bf.shape[0]), interpolation=cv.INTER_CUBIC)

# Muestra las imgenes redimensionadas
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs[0, 0].imshow(cv.cvtColor(img1_resized_nearest, cv.COLOR_BGR2RGB))
axs[0, 0].set_title("INTER_NEAREST")
axs[0, 1].imshow(cv.cvtColor(img1_resized_cubic, cv.COLOR_BGR2RGB))
axs[0, 1].set_title("INTER_CUBIC")
axs[1, 0].imshow(cv.cvtColor(img2_resized_nearest, cv.COLOR_BGR2RGB))
axs[1, 0].set_title("INTER_NEAREST")
axs[1, 1].imshow(cv.cvtColor(img2_resized_cubic, cv.COLOR_BGR2RGB))
axs[1, 1].set_title("INTER_CUBIC")
plt.show()
```

```python
# Redimensiona las imgenes modificando solo el factor de escala en el eje verticial a 2
img1_resized_nearest = cv.resize(imag_halo, (imag_halo.shape[1], imag_halo.shape[0]*2), interpolation=cv.INTER_NEAREST)
img1_resized_cubic = cv.resize(imag_halo, (imag_halo.shape[1], imag_halo.shape[0]*2), interpolation=cv.INTER_CUBIC)
img2_resized_nearest = cv.resize(imag_bf, (imag_bf.shape[1], imag_bf.shape[0]*2), interpolation=cv.INTER_NEAREST)
img2_resized_cubic = cv.resize(imag_bf, (imag_bf.shape[1], imag_bf.shape[0]*2), interpolation=cv.INTER_CUBIC)

# Muestra las imgenes redimensionadas
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs[0, 0].imshow(cv.cvtColor(img1_resized_nearest, cv.COLOR_BGR2RGB))
axs[0, 0].set_title("INTER_NEAREST")
axs[0, 1].imshow(cv.cvtColor(img1_resized_cubic, cv.COLOR_BGR2RGB))
axs[0, 1].set_title("INTER_CUBIC")
axs[1, 0].imshow(cv.cvtColor(img2_resized_nearest, cv.COLOR_BGR2RGB))
axs[1, 0].set_title("INTER_NEAREST")
axs[1, 1].imshow(cv.cvtColor(img2_resized_cubic, cv.COLOR_BGR2RGB))
axs[1, 1].set_title("INTER_CUBIC")
plt.show()
```

```python
height, width = imag_halo.shape[:2]
M = np.float32([[1, 0, 50], [0, 1, 0]])
imag1_desplazada = cv.warpAffine(imag_halo, M, (width, height))

fig, axs = plt.subplots(1, 2, figsize=(8, 8))

axs[0].imshow(cv.cvtColor(imag_halo, cv.COLOR_BGR2RGB))
axs[0].set_title('Imagen 1 Original')
axs[0].axis('off')

axs[1].imshow(cv.cvtColor(imag1_desplazada, cv.COLOR_BGR2RGB))
axs[1].set_title('Imagen 1 Desplazada')
axs[1].axis('off')
```

```python
M.shape
```

```python
angle = 45
M = cv.getRotationMatrix2D((width/2, height/2), angle, 1)

# Aplica la rotacin a la imagen original
rotated_img = cv.warpAffine(imag_halo, M, (width, height))

# Muestra la imagen original y la imagen rotada en un subplot
plt.subplot(121)
plt.imshow(imag_halo[:,:,::-1])
plt.title('Imagen original')

plt.subplot(122)
plt.imshow(rotated_img[:,:,::-1])
plt.title('Imagen rotada 45 grados')

plt.show()
```
